# Exercise 1.2 :  K-Fold Cross-Validation with Keras

n this exercise, I implement K-Fold Cross-Validation to evaluate the performance of a neural network model.

Instead of splitting the dataset into a single training and test set, K-Fold Cross-Validation divides the dataset into K subsets (folds). The model is then trained and evaluated K times, each time using a different fold as the validation set and the remaining folds for training

## importing libruaries :

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from scikeras.wrappers import KerasClassifier

from sklearn.model_selection import cross_val_score

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from sklearn.model_selection import train_test_split
```

## splitting data :


```python
X_train , X_test , y_train , y_test = train_test_split (X , y ,
test_size =0.2 , random_state =0)
```

## Feature scaling :

```python
sc = StandardScaler ()
 X_train = sc . fit_transform ( X_train )
 X_test = sc . transform ( X_test )
```

## Building the Neural Network :

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units=6, activation='relu', input_dim=11))
    classifier.add(Dense(units=6, activation='relu'))
    classifier.add(Dense(units=1, activation='sigmoid'))
    classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return classifier
```

## creating classifier :


```python
classifier = KerasClassifier(
    build_fn=build_classifier,
    batch_size=10,
    epochs=100,
    verbose=0
) # <-- returns raw Keras model
```

## performing k-Fold cross-validation :

```python
accuracies = cross_val_score ( estimator = classifier , X = X_train , y =
y_train , cv =10)
```
