# Exercise 1.1: Binary Classification with Keras

## importing libraries :


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

## Loading dataset :

```python
dataset = pd.read_csv('Churn_Modelling.csv')

X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

## seeing and understanding the datset :

```python
print(dataset.info()) #nb rows columns dta type missing values
```

all columns have 10000 non-null count => no missing values 
our target => exited (o / 1)

## encoding categorical data :


```python
le1 = LabelEncoder()
X[:,1] = le1.fit_transform(X[:,1])

le2 = LabelEncoder()
X[:,2] = le2.fit_transform(X[:,2])
```

## One-Hot Encoding :


```python
ct = ColumnTransformer([("Geography", OneHotEncoder(), [1])],
                       remainder='passthrough')

X = ct.fit_transform(X)
X = X[:,1:]
```

## Train Test Split :

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)
```

## Feature Scaling :

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Building Neural Network :

```python
classifier = Sequential()

classifier.add(Dense(units=6, activation='relu', input_dim=11))
classifier.add(Dense(units=6, activation='relu'))

classifier.add(Dropout(0.2))

classifier.add(Dense(units=1, activation='sigmoid'))
```

```python
from keras.layers import Input

classifier = Sequential([
    Input(shape=(11,)),
    Dense(6, activation='relu'),
    Dense(6, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])
```

## Compiling the model :

```python
classifier.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

## Training the model :

```python
classifier.fit(X_train, y_train,
               batch_size=10,
               epochs=100)
```

## Prediction :

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

## Confusion matrix :

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```
