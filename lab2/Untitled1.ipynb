```python
# =========================
# 1. IMPORT LIBRARIES
# =========================
import numpy as np
import pandas as pd

# dataset
from sklearn.datasets import load_boston

# preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# classical models
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor

# neural network
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# =========================
# 2. LOAD DATA
# =========================
data = load_boston()

X = data.data      # features
y = data.target    # house prices

# =========================
# 3. TRAIN / TEST SPLIT
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

# =========================
# 4. SCALING
# =========================
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# =========================
# 5. NEURAL NETWORK MODEL
# =========================
model = Sequential()

# hidden layers
model.add(Dense(32, activation="relu", input_dim=X_train.shape[1]))
model.add(Dense(16, activation="relu"))

# output layer (regression → 1 value, no sigmoid!)
model.add(Dense(1))

# compile
model.compile(
    optimizer="adam",
    loss="mse"   # regression uses MSE
)

# train
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)

# predict
y_pred_nn = model.predict(X_test)

# =========================
# 6. EVALUATION FUNCTION
# =========================
def evaluate(name, y_true, y_pred):
    print(f"\n{name}")
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("R2 :", r2_score(y_true, y_pred))

# evaluate NN
evaluate("Neural Network", y_test, y_pred_nn)

# =========================
# 7. CLASSICAL MODELS
# =========================

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
evaluate("Linear Regression", y_test, y_pred_lr)

# Ridge
ridge = Ridge()
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
evaluate("Ridge", y_test, y_pred_ridge)

# Lasso
lasso = Lasso()
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
evaluate("Lasso", y_test, y_pred_lasso)

# Random Forest
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
evaluate("Random Forest", y_test, y_pred_rf)
```
