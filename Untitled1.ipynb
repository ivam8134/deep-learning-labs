```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## loading dataset 

```python

data_url = "http://lib.stat.cmu.edu/datasets/boston"

raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# Build features and target
X = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y = raw_df.values[1::2, 2]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

Dataset loaded manually (since removed from sklearn)
506 samples, 13 features

## exploring data

```python

df = pd.DataFrame(X)
df["PRICE"] = y

df.head()
```

## splitting

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

## feature scaling


```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## results storage

```python
results = []
histories = {}
predictions = {}
```

## basic neural network

- 2 hidden layers
- ReLU activation
- Output = 1 (regression)

## deeper nearal network

- More layers → more complex model
- May improve or overfit

```python
def train_model(name, activation):

    model = Sequential()

    # Input layer + hidden layers
    model.add(Dense(64, activation=activation, input_dim=X_train.shape[1]))
    model.add(Dense(32, activation=activation))
    model.add(Dense(16, activation=activation))

    # Output layer
    model.add(Dense(1))

    model.compile(optimizer="adam", loss="mse")

    history = model.fit(
        X_train, y_train,
        epochs=120,
        batch_size=32,
        verbose=0
    )

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"\nModel: {name}")
    print("MSE:", mse)
    print("MAE:", mae)
    print("R2:", r2)

    results.append([name, mse, mae, r2])
    histories[name] = history
    predictions[name] = y_pred
```

```python
train_model("ReLU", "relu")
train_model("tanh", "tanh")
train_model("sigmoid", "sigmoid")
train_model("elu", "elu")
train_model("selu", "selu")
```

## model comparison

```python
df = pd.DataFrame(results, columns=["Model", "MSE", "MAE", "R2"])

plt.figure()
plt.bar(df["Model"], df["R2"])
plt.title("Model Comparison (R2 Score)")
plt.ylabel("R2")
plt.xticks(rotation=45)
plt.show()
```

## loss curves

```python
plt.figure()

for name, history in histories.items():
    plt.plot(history.history["loss"], label=name)

plt.title("Training Loss Comparison")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

## real vs predicted (best model)

```python
best_model = df.sort_values("R2", ascending=False).iloc[0]["Model"]

y_pred_best = predictions[best_model]

plt.figure()
plt.scatter(y_test, y_pred_best)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title(f"Actual vs Predicted ({best_model})")
plt.show()
```
