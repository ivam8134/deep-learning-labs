# Exercice 1

```python

import numpy as np
import matplotlib.pyplot as plt


import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, datasets

np.random.seed(42)
tf.random.set_seed(42)
```

```python
(x_train_raw, y_train), (x_test_raw, y_test) = datasets.mnist.load_data()

print("=== Raw data shapes ===")
print(f"  x_train : {x_train_raw.shape}" )
print(f"  x_test  : {x_test_raw.shape}")
print(f"  y_train : {y_train.shape} ")
print(f"  Pixel value range before normalisation: [{x_train_raw.min()}, {x_train_raw.max()}]")
```

```python
x_train = x_train_raw.astype('float32') / 255.0
x_test  = x_test_raw.astype('float32')  / 255.0

x_train = x_train.reshape(-1, 28, 28, 1)  # -1 means "keep the number of images as-is"
x_test  = x_test.reshape(-1, 28, 28, 1)

print("\n=== After preprocessing ===")
print(f"  x_train : {x_train.shape}")
print(f"  x_test  : {x_test.shape}")
print(f"  Pixel value range after normalisation: [{x_train.min():.1f}, {x_train.max():.1f}]")
```

```python
# Visualise a few samples 
fig, axes = plt.subplots(2, 10, figsize=(16, 4))
fig.suptitle('MNIST samples — one per class', fontsize=14, fontweight='bold')

for digit in range(10):
    # Find the first image belonging to class `digit`
    idx = np.where(y_train == digit)[0][0]
    axes[0, digit].imshow(x_train[idx, :, :, 0], cmap='gray')
    axes[0, digit].set_title(f'Label: {digit}', fontsize=9)
    axes[0, digit].axis('off')

    # Show a second example of the same class
    idx2 = np.where(y_train == digit)[0][1]
    axes[1, digit].imshow(x_train[idx2, :, :, 0], cmap='gray')
    axes[1, digit].axis('off')

plt.tight_layout()
plt.show()
```

```python
def build_cnn_mnist():
    """
    Build the CNN for MNIST as specified in Exercise 1.
    Returns a compiled Keras model.
    """
    model = models.Sequential(name="CNN_MNIST")  # Sequential = layers stacked one after another

    # ── BLOCK 1: First convolutional block ───────────────────────────────────
    # 32 filters of size 3×3.  padding='valid' means NO padding → output shrinks.
    # Input: (28, 28, 1)   Output: (26, 26, 32)   because (28-3+1)=26
    model.add(layers.Conv2D(
        filters=32,
        kernel_size=(3, 3),
        activation='relu',          # Apply ReLU right after convolution
        input_shape=(28, 28, 1),    # Must specify input shape in the FIRST layer only
        name='conv1'
    ))

    # MaxPooling: take the max in each 2×2 window  →  halves width & height
    # Input: (26, 26, 32)   Output: (13, 13, 32)
    model.add(layers.MaxPooling2D(pool_size=(2, 2), name='pool1'))

    # ── BLOCK 2: Second convolutional block ──────────────────────────────────
    # Deeper layers use MORE filters (64) because they detect more abstract patterns.
    # Input: (13, 13, 32)   Output: (11, 11, 64)
    model.add(layers.Conv2D(
        filters=64,
        kernel_size=(3, 3),
        activation='relu',
        name='conv2'
    ))

    # Input: (11, 11, 64)   Output: (5, 5, 64)   (floor(11/2) = 5)
    model.add(layers.MaxPooling2D(pool_size=(2, 2), name='pool2'))

    # ── FLATTEN: 3D → 1D ─────────────────────────────────────────────────────
    # (5, 5, 64)  →  (1600,)
    model.add(layers.Flatten(name='flatten'))

    # ── DENSE LAYERS: classification head ────────────────────────────────────
    model.add(layers.Dense(128, activation='relu', name='dense1'))

    # Final layer: 10 neurons (one per digit), softmax turns logits into probabilities
    model.add(layers.Dense(10, activation='softmax', name='output'))

    return model


cnn_model = build_cnn_mnist()
cnn_model.summary()
```

```python
cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  
    metrics=['accuracy']
)
```

```python
print("Training CNN on MNIST for 10 epochs …")
history_cnn = cnn_model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=64,         # Process 64 images at a time
    validation_split=0.1, 
    verbose=1              
)
```

```python
test_loss, test_acc = cnn_model.evaluate(x_test, y_test, verbose=0)
print(f"\n CNN Test Accuracy  : {test_acc * 100:.2f}%")
print(f"   CNN Test Loss      : {test_loss:.4f}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Exercise 1 — CNN on MNIST: Learning Curves', fontsize=14, fontweight='bold')

epochs_range = range(1, 11)

# Accuracy plot
ax1.plot(epochs_range, history_cnn.history['accuracy'],     'b-o', label='Train Accuracy')
ax1.plot(epochs_range, history_cnn.history['val_accuracy'], 'r--s', label='Val Accuracy')
ax1.set_title('Accuracy per Epoch')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0.9, 1.01])

# Loss plot
ax2.plot(epochs_range, history_cnn.history['loss'],     'b-o', label='Train Loss')
ax2.plot(epochs_range, history_cnn.history['val_loss'], 'r--s', label='Val Loss')
ax2.set_title('Loss per Epoch')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

# Exercice 2

```python
(x_fash_train_raw, y_fash_train), (x_fash_test_raw, y_fash_test) = datasets.fashion_mnist.load_data()

class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
               'Sandal','Shirt','Sneaker','Bag','Ankle boot']
```

```python
#  Normalise 
x_fash_train = x_fash_train_raw.astype('float32') / 255.0
x_fash_test  = x_fash_test_raw.astype('float32')  / 255.0

print(f"Training set: {x_fash_train.shape}  |  Test set: {x_fash_test.shape}")
```

```python
fig, axes = plt.subplots(2, 10, figsize=(18, 4))
fig.suptitle('Fashion-MNIST samples — one per class', fontsize=13, fontweight='bold')
for c in range(10):
    idx = np.where(y_fash_train == c)[0][0]
    axes[0, c].imshow(x_fash_train[idx], cmap='gray')
    axes[0, c].set_title(class_names[c], fontsize=7)
    axes[0, c].axis('off')
    idx2 = np.where(y_fash_train == c)[0][5]
    axes[1, c].imshow(x_fash_train[idx2], cmap='gray')
    axes[1, c].axis('off')
plt.tight_layout()
plt.show()
```

```python
def build_ann_fashion():
 
    model = models.Sequential(name="ANN_Fashion")

    model.add(layers.Flatten(input_shape=(28, 28), name='flatten'))

    model.add(layers.Dense(256, activation='relu', name='hidden'))

    model.add(layers.Dense(10, activation='softmax', name='output'))

    return model


ann_model = build_ann_fashion()
ann_model.summary()
```

```python
def build_cnn_fashion():
    
    model = models.Sequential(name="CNN_Fashion")

    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), name='conv1'))
    model.add(layers.MaxPooling2D((2, 2), name='pool1'))

    model.add(layers.Conv2D(64, (3, 3), activation='relu', name='conv2'))
    model.add(layers.MaxPooling2D((2, 2), name='pool2'))

    model.add(layers.Flatten(name='flatten'))
    model.add(layers.Dense(128, activation='relu', name='dense1'))
    model.add(layers.Dense(10, activation='softmax', name='output'))

    return model


cnn_fashion = build_cnn_fashion()
cnn_fashion.summary()
```

```python
ann_params = ann_model.count_params()
cnn_params = cnn_fashion.count_params()

print(f"{'ANN':<20} {ann_params:>20,}")
print(f"{'CNN':<20} {cnn_params:>20,}")

print(f"  The ANN has {ann_params / cnn_params:.1f}× MORE parameters than the CNN!")
print("  Yet the CNN will achieve HIGHER accuracy — because of spatial inductive bias.")
```

```python
ann_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

print("Training ANN on Fashion-MNIST")
history_ann = ann_model.fit(
    x_fash_train, y_fash_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1,
    verbose=1
)
```

```python
#compile and train CNN
x_fash_train_cnn = x_fash_train.reshape(-1, 28, 28, 1)
x_fash_test_cnn  = x_fash_test.reshape(-1, 28, 28, 1)

cnn_fashion.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

print("Training CNN on Fashion-MNIST …")
history_cnn_fashion = cnn_fashion.fit(
    x_fash_train_cnn, y_fash_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1,
    verbose=1
)
```

```python
ann_loss, ann_acc = ann_model.evaluate(x_fash_test, y_fash_test, verbose=0)
cnn_loss, cnn_acc = cnn_fashion.evaluate(x_fash_test_cnn, y_fash_test, verbose=0)

print(f"{'Model':<12} {'Test Accuracy':>15} {'Parameters':>15}")
print("=" * 55)
print(f"{'ANN':<12} {ann_acc*100:>14.2f}% {ann_params:>15,}")
print(f"{'CNN':<12} {cnn_acc*100:>14.2f}% {cnn_params:>15,}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Exercise 2 — ANN vs CNN on Fashion-MNIST', fontsize=14, fontweight='bold')
epochs_range = range(1, 11)

for ax, metric, title in zip(axes, ['accuracy', 'loss'], ['Accuracy', 'Loss']):
    ax.plot(epochs_range, history_ann.history[f'val_{metric}'],
            'r--o', label=f'ANN val_{metric}', linewidth=2)
    ax.plot(epochs_range, history_cnn_fashion.history[f'val_{metric}'],
            'b-s', label=f'CNN val_{metric}', linewidth=2)
    ax.set_title(f'Validation {title}')
    ax.set_xlabel('Epoch')
    ax.set_ylabel(metric.capitalize())
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

# Exercice 3

```python
(x_cifar_train, y_cifar_train), (x_cifar_test, _) = datasets.cifar10.load_data()

cifar_classes = ['airplane','automobile','bird','cat','deer',
                 'dog','frog','horse','ship','truck']
```

```python
x_cifar_train = x_cifar_train.astype('float32') / 255.0

print(f"CIFAR-10 training set shape: {x_cifar_train.shape}")
```

```python
sample_imgs = x_cifar_train[:8]
sample_labels = y_cifar_train[:8].flatten()

fig, axes = plt.subplots(1, 8, figsize=(18, 3))
fig.suptitle('CIFAR-10 — 8 sample images', fontsize=13, fontweight='bold')
for i, ax in enumerate(axes):
    ax.imshow(sample_imgs[i])
    ax.set_title(cifar_classes[sample_labels[i]], fontsize=9)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
batch = x_cifar_train[:8]  # Shape: (8, 32, 32, 3)

configurations = [
    ('valid', 1),
    ('valid', 2),
    ('same',  1),
    ('same',  2),
]

results = {}  # for visualisation

for padding, stride in configurations:
    
    conv_layer = layers.Conv2D(
        filters=1,
        kernel_size=(3, 3),
        strides=(stride, stride),
        padding=padding,
        activation='relu',
        use_bias=False
    )

    output = conv_layer(batch)   
    h_out, w_out = output.shape[1], output.shape[2]

    # Compute expected size using the formula
    if padding == 'valid':
        expected = (32 - 3) // stride + 1
        formula = f"(32-3+0)/{stride}+1 = {expected}"
    else:  # same
        import math
        expected = math.ceil(32 / stride)
        formula = f"ceil(32/{stride}) = {expected}"

    key = f"padding={padding}, stride={stride}"
    results[key] = output.numpy()

    print(f"{padding:<10} {stride:<8} {str(output.shape):<22} {formula}")
```

```python

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
fig.suptitle('Exercise 3 — Original image & feature maps for each (padding, stride) combo',
             fontsize=12, fontweight='bold')

# Original image
axes[0].imshow(batch[0])   
axes[0].set_title(f'Original\n32×32×3', fontsize=10, fontweight='bold')
axes[0].axis('off')


for ax, (key, output) in zip(axes[1:], results.items()):
    fm = output[0, :, :, 0]   # First image, first (only) filter
    ax.imshow(fm, cmap='viridis')
    h, w = fm.shape
    ax.set_title(f'{key}\n{h}×{w}', fontsize=9)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

# Exercice 4

```python
sample_img = x_cifar_train[0:1]   # shape (1, 32, 32, 3)

augment_demo = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),    # ±10% of 360° ≈ ±36°
], name="augmentation_demo")

fig, axes = plt.subplots(2, 8, figsize=(18, 5))
fig.suptitle('Data Augmentation — same image transformed differently each time',
             fontsize=13, fontweight='bold')
axes[0, 0].imshow(sample_img[0])
axes[0, 0].set_title('Original', fontweight='bold', fontsize=9)
axes[0, 0].axis('off')

for i in range(1, 16):
    ax = axes[i // 8, i % 8]
    aug = augment_demo(sample_img, training=True)   # training=True applies randomness
    ax.imshow(aug[0].numpy())
    ax.set_title(f'Aug #{i}', fontsize=8)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
