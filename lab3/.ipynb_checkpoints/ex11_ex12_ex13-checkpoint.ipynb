# Exercice 11

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import time
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers, losses, callbacks
from tensorflow.keras.applications import VGG16, ResNet50, InceptionV3
from tensorflow.keras.datasets import cifar10, fashion_mnist
from tensorflow.keras.utils import to_categorical


tf.random.set_seed(42)
np.random.seed(42)
```

```python
(x_train_full, y_train_full), (x_test_full, y_test_full) = cifar10.load_data()
def filter_cats_dogs(x, y):
    mask = (y.flatten() == 3) | (y.flatten() == 5)   # cat or dog
    x_f = x[mask].astype('float32') / 255.0           # normalize to [0,1]
    y_f = (y[mask].flatten() == 5).astype('float32')  # dog=1, cat=0
    return x_f, y_f

x_train_cd, y_train_cd = filter_cats_dogs(x_train_full, y_train_full)
x_test_cd,  y_test_cd  = filter_cats_dogs(x_test_full,  y_test_full)
```

```python

print(f"Training samples : {x_train_cd.shape[0]}")
print(f"Test samples     : {x_test_cd.shape[0]}")
print(f"Image shape      : {x_train_cd.shape[1:]}")

# Show a few examples
fig, axes = plt.subplots(1, 8, figsize=(16, 2))
labels = ['cat', 'dog']
for i, ax in enumerate(axes):
    ax.imshow(x_train_cd[i])
    ax.set_title(labels[int(y_train_cd[i])], fontsize=9)
    ax.axis('off')
plt.suptitle('Sample images – Cats vs Dogs', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
def build_scratch_model(input_shape=(32, 32, 3)):
  
    model = models.Sequential([
        layers.Input(shape=input_shape),

        # Block 1 — learn low-level features (edges, colours)
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),   # 32×32 → 16×16

        # Block 2 — mid-level features (textures, parts)
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),   # 16×16 → 8×8

        # Block 3 — higher-level features (object parts)
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),   # 8×8 → 4×4

        # Classification head
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),        
        layers.Dense(1, activation='sigmoid')  
    ], name='scratch_model')
    return model

scratch_model = build_scratch_model()
scratch_model.summary()


scratch_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
print("Training CNN from scratch...")
t0 = time.time()

history_scratch = scratch_model.fit(
    x_train_cd, y_train_cd,
    epochs=20,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

scratch_time = time.time() - t0
scratch_acc  = scratch_model.evaluate(x_test_cd, y_test_cd, verbose=0)[1]
print(f"\nScratch model — Test accuracy: {scratch_acc:.4f} | Time: {scratch_time:.1f}s")
```

```python
conv_base = VGG16(weights='imagenet', include_top=False,
                  input_shape=(32, 32, 3))


conv_base.trainable = False

print(f"VGG16 conv base parameters : {conv_base.count_params():,}")
print(f"Trainable parameters       : {sum([tf.size(v).numpy() for v in conv_base.trainable_variables]):,}")
```

```python
def build_transfer_model(conv_base):
    model = models.Sequential([
        conv_base,
        layers.GlobalAveragePooling2D(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')   
    ], name='transfer_model')
    return model

transfer_model = build_transfer_model(conv_base)
transfer_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
transfer_model.summary()

print("\nTraining transfer learning model...")
t0 = time.time()

history_transfer = transfer_model.fit(
    x_train_cd, y_train_cd,
    epochs=20,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

transfer_time = time.time() - t0
transfer_acc  = transfer_model.evaluate(x_test_cd, y_test_cd, verbose=0)[1]
print(f"\nTransfer model — Test accuracy: {transfer_acc:.4f} | Time: {transfer_time:.1f}s")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

for ax, h_scratch, h_transfer, metric, title in [
    (ax1, history_scratch.history['val_accuracy'],
          history_transfer.history['val_accuracy'],
          'accuracy', 'Validation Accuracy'),
    (ax2, history_scratch.history['val_loss'],
          history_transfer.history['val_loss'],
          'loss', 'Validation Loss')
]:
    ax.plot(h_scratch,   label='From Scratch', color='tomato',  linewidth=2)
    ax.plot(h_transfer,  label='Transfer (VGG16)', color='steelblue', linewidth=2)
    ax.set_title(title, fontsize=13, fontweight='bold')
    ax.set_xlabel('Epoch')
    ax.legend()
    ax.grid(alpha=0.3)

plt.suptitle('Exercise 11 — Scratch vs Transfer Learning', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
print("VGG16 layers:")
for i, layer in enumerate(conv_base.layers):
    print(f"  [{i:2d}] {layer.name:<20s}  trainable={layer.trainable}")
```

```python

conv_base.trainable = True  # enable updates for the whole base

FREEZE_UNTIL = len(conv_base.layers) - 4
for layer in conv_base.layers[:FREEZE_UNTIL]:
    layer.trainable = False

print("Layers after fine-tune setup:")
for i, layer in enumerate(conv_base.layers):
    status = ' trainable' if layer.trainable else ' frozen'
    print(f"  [{i:2d}] {layer.name:<20s}  {status}")
```

```python
transfer_model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),  # 100× smaller than default
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print(f"\nTrainable parameters after unfreezing: {transfer_model.count_params():,}")
```

```python
print("Fine-tuning...")
history_finetune = transfer_model.fit(
    x_train_cd, y_train_cd,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

finetune_acc = transfer_model.evaluate(x_test_cd, y_test_cd, verbose=0)[1]
print(f"\nAfter fine-tuning — Test accuracy: {finetune_acc:.4f}")
print(f"Before fine-tuning — Test accuracy: {transfer_acc:.4f}")
```

```python
plt.figure(figsize=(10, 4))
plt.plot(history_finetune.history['accuracy'],     label='Train', color='steelblue')
plt.plot(history_finetune.history['val_accuracy'], label='Val',   color='orange', linestyle='--')
plt.title('Exercise 12 — Fine-Tuning Accuracy', fontsize=13, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# Exercice 13

```python
(x_train_c10, y_train_c10), (x_test_c10, y_test_c10) = cifar10.load_data()
x_train_c10 = x_train_c10.astype('float32') / 255.0
x_test_c10  = x_test_c10.astype('float32')  / 255.0
y_train_c10 = to_categorical(y_train_c10, 10)
y_test_c10  = to_categorical(y_test_c10,  10)
```

```python

def build_standard_cnn(input_shape=(32, 32, 3)):
   
    return models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),
        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(10,  activation='softmax')
    ], name='standard_conv')
```

```python
def build_separable_cnn(input_shape=(32, 32, 3)):
    return models.Sequential([
        layers.Input(shape=input_shape),
        # SeparableConv2D is a drop-in replacement for Conv2D
        layers.SeparableConv2D(32, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),
        layers.SeparableConv2D(64, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(10,  activation='softmax')
    ], name='separable_conv')
```

```python
std_model = build_standard_cnn()
sep_model = build_separable_cnn()

std_params = std_model.count_params()
sep_params = sep_model.count_params()

print(f"Standard Conv  — Total params: {std_params:,}")
print(f"Separable Conv — Total params: {sep_params:,}")
print(f"Parameter reduction: {std_params/sep_params:.2f}× fewer params with SeparableConv")
```

```python
for name, model in [('Standard Conv', std_model), ('Separable Conv', sep_model)]:
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("Training Standard Conv...")
t0 = time.time()
h_std = std_model.fit(x_train_c10, y_train_c10, epochs=10, batch_size=64,
                       validation_split=0.1, verbose=0)
std_time = time.time() - t0
std_acc  = std_model.evaluate(x_test_c10, y_test_c10, verbose=0)[1]

print("Training Separable Conv...")
t0 = time.time()
h_sep = sep_model.fit(x_train_c10, y_train_c10, epochs=10, batch_size=64,
                       validation_split=0.1, verbose=0)
sep_time = time.time() - t0
sep_acc  = sep_model.evaluate(x_test_c10, y_test_c10, verbose=0)[1]
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(h_std.history['val_accuracy'], label='Standard Conv', color='tomato',    linewidth=2)
ax.plot(h_sep.history['val_accuracy'], label='Separable Conv', color='steelblue', linewidth=2)
ax.set_title('Exercise 13 — Standard vs Separable Convolution', fontsize=13, fontweight='bold')
ax.set_xlabel('Epoch'); ax.set_ylabel('Val Accuracy')
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

print(f"\n📊 Standard Conv  — Acc: {std_acc:.4f}  | Params: {std_params:,}  | Time: {std_time:.1f}s")
print(f"   Separable Conv — Acc: {sep_acc:.4f}  | Params: {sep_params:,}  | Time: {sep_time:.1f}s")
```
